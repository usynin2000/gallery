## –í—Å–ø–æ–º–æ–≥–∞—Ç–µ–ª—å–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è –¥–ª—è –±—ã—Å—Ç—Ä–æ–≥–æ —Ä–µ–Ω–¥–µ—Ä–∏–Ω–≥–∞

–≠—Ç–∞ —Ñ—É–Ω–∫—Ü–∏—è –ø–æ–∑–≤–æ–ª—è–µ—Ç —Å–æ–∑–¥–∞—Ç—å –∏ —Å—Ä–∞–∑—É —É–≤–∏–¥–µ—Ç—å –∞–Ω–∏–º–∞—Ü–∏—é:


```python
def render_scene(scene_class, quality="low_quality"):
    """
    –†–µ–Ω–¥–µ—Ä–∏—Ç —Å—Ü–µ–Ω—É –∏ –æ—Ç–æ–±—Ä–∞–∂–∞–µ—Ç —Ä–µ–∑—É–ª—å—Ç–∞—Ç –ø—Ä—è–º–æ –≤ notebook.
    
    –ü–∞—Ä–∞–º–µ—Ç—Ä—ã:
    - scene_class: –∫–ª–∞—Å—Å —Å—Ü–µ–Ω—ã –¥–ª—è —Ä–µ–Ω–¥–µ—Ä–∏–Ω–≥–∞
    - quality: 'low_quality' (–±—ã—Å—Ç—Ä–æ), 'medium_quality', 'high_quality' (–º–µ–¥–ª–µ–Ω–Ω–æ)
    """
    # –í—Ä–µ–º–µ–Ω–Ω–æ –º–µ–Ω—è–µ–º –∫–∞—á–µ—Å—Ç–≤–æ
    old_quality = config.quality
    config.quality = quality
    
    # –†–µ–Ω–¥–µ—Ä–∏–º —Å—Ü–µ–Ω—É (–ª–æ–≥–∏ INFO –±—É–¥—É—Ç —Å–∫—Ä—ã—Ç—ã –±–ª–∞–≥–æ–¥–∞—Ä—è –Ω–∞—Å—Ç—Ä–æ–π–∫–µ –≤—ã—à–µ)
    scene = scene_class()
    scene.render()
    
    # –í–æ—Å—Å—Ç–∞–Ω–∞–≤–ª–∏–≤–∞–µ–º –∫–∞—á–µ—Å—Ç–≤–æ
    config.quality = old_quality
    
    # –û—Ç–æ–±—Ä–∞–∂–∞–µ–º –≤–∏–¥–µ–æ
    video_path = scene.renderer.file_writer.movie_file_path
    return Video(video_path, embed=True, width=600)
```

## üéØ –ü—Ä–∏–º–µ—Ä 1: –ë–∞–∑–æ–≤–∞—è –∞–Ω–∏–º–∞—Ü–∏—è - –ö–≤–∞–¥—Ä–∞—Ç –ø—Ä–µ–≤—Ä–∞—â–∞–µ—Ç—Å—è –≤ –∫—Ä—É–≥


```python
class SquareToCircle(Scene):
    def construct(self):
        # –°–æ–∑–¥–∞–µ–º –∫—Ä—É–≥
        circle = Circle()
        circle.set_fill(PINK, opacity=0.5)
        
        # –°–æ–∑–¥–∞–µ–º –∫–≤–∞–¥—Ä–∞—Ç
        square = Square()
        square.set_fill(BLUE, opacity=0.5)
        square.rotate(PI / 4)
        
        # –ê–Ω–∏–º–∞—Ü–∏–∏
        self.play(Create(square))
        self.wait(0.5)
        self.play(Transform(square, circle))
        self.wait(0.5)
        self.play(FadeOut(square))

# –†–µ–Ω–¥–µ—Ä–∏–º –∏ –ø–æ–∫–∞–∑—ã–≤–∞–µ–º
render_scene(SquareToCircle)
```

## üéØ –ü—Ä–∏–º–µ—Ä 3: –ê–Ω–∏–º–∞—Ü–∏—è –≥—Ä–∞—Ñ–∏–∫–∞ —Ñ—É–Ω–∫—Ü–∏–∏


```python
class PlotFunction(Scene):
    def construct(self):
        # –°–æ–∑–¥–∞–µ–º –æ—Å–∏
        axes = Axes(
            x_range=[-3, 3, 1],
            y_range=[-3, 3, 1],
            x_length=6,
            y_length=6,
            axis_config={"color": BLUE},
        )
        
        # –ü–æ–¥–ø–∏—Å–∏ –æ—Å–µ–π
        labels = axes.get_axis_labels(x_label="x", y_label="f(x)")
        
        # –ì—Ä–∞—Ñ–∏–∫ —Ñ—É–Ω–∫—Ü–∏–∏
        graph = axes.plot(lambda x: x**2, color=GREEN)
        graph_label = axes.get_graph_label(graph, label="x^2")
        
        # –ê–Ω–∏–º–∞—Ü–∏–∏
        self.play(Create(axes), Write(labels))
        self.wait(0.5)
        self.play(Create(graph), run_time=2)
        self.play(Write(graph_label))
        self.wait(1)

render_scene(PlotFunction)
```

## üéØ –ü—Ä–∏–º–µ—Ä 4: –ê–Ω–∏–º–∞—Ü–∏—è —Å –Ω–µ—Å–∫–æ–ª—å–∫–∏–º–∏ –æ–±—ä–µ–∫—Ç–∞–º–∏


```python
class MultipleShapes(Scene):
    def construct(self):
        # –°–æ–∑–¥–∞–µ–º –Ω–µ—Å–∫–æ–ª—å–∫–æ —Ñ–∏–≥—É—Ä
        circle = Circle(radius=0.5, color=RED)
        square = Square(side_length=1, color=BLUE)
        triangle = Triangle(color=GREEN)
        
        # –†–∞–∑–º–µ—â–∞–µ–º –∏—Ö
        circle.shift(LEFT * 3)
        triangle.shift(RIGHT * 3)
        
        # –ì—Ä—É–ø–ø–∏—Ä—É–µ–º
        shapes = VGroup(circle, square, triangle)
        
        # –ê–Ω–∏–º–∞—Ü–∏–∏
        self.play(FadeIn(shapes, shift=UP))
        self.wait(0.5)
        
        # –í—Ä–∞—â–µ–Ω–∏–µ –≤—Å–µ—Ö —Ñ–∏–≥—É—Ä
        self.play(Rotate(shapes, angle=PI, run_time=2))
        self.wait(0.5)
        
        # –ú–∞—Å—à—Ç–∞–±–∏—Ä–æ–≤–∞–Ω–∏–µ
        self.play(shapes.animate.scale(1.5))
        self.wait(0.5)
        
        # –ò–∑–º–µ–Ω–µ–Ω–∏–µ —Ü–≤–µ—Ç–∞
        self.play(
            circle.animate.set_fill(YELLOW, opacity=0.7),
            square.animate.set_fill(PINK, opacity=0.7),
            triangle.animate.set_fill(ORANGE, opacity=0.7)
        )
        self.wait(1)

render_scene(MultipleShapes)
```

## üéØ –ü—Ä–∏–º–µ—Ä 5: –ê–Ω–∏–º–∏—Ä–æ–≤–∞–Ω–Ω–∞—è —Ç—Ä–∞–Ω—Å—Ñ–æ—Ä–º–∞—Ü–∏—è —Å —á–∏—Å–ª–∞–º–∏


```python
class NumberTransformation(Scene):
    def construct(self):
        # –ù–∞—á–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ
        number = DecimalNumber(0, num_decimal_places=2)
        number.scale(2)
        
        self.add(number)
        self.wait(0.5)
        
        # –ê–Ω–∏–º–∞—Ü–∏—è –∏–∑–º–µ–Ω–µ–Ω–∏—è —á–∏—Å–ª–∞
        self.play(
            number.animate.set_value(100),
            run_time=3,
            rate_func=smooth
        )
        self.wait(1)

render_scene(NumberTransformation)
```

–ò—Å–ø–æ–ª—å–∑—É–π—Ç–µ —ç—Ç—É —è—á–µ–π–∫—É –¥–ª—è —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–æ–≤:
